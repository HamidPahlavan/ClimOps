```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
# load dataset
df = pd.read_csv('../../us-census-demographic-data/acs2015_county_data.csv')
df.head()

list(df)
```

```python
# get unique values in "State" 
print(set(df['State']))
```

```python
# how many counties does each state have
county = df[['State','County']].groupby('State').agg('count')
county.sort_values('County', ascending = 0)
```

```python
#Aggregate Data By State
# group by state 
agg_df = df.groupby('State').sum()

# order by total population 
agg_df.sort_values('TotalPop', ascending = 0)
```

```python
fig = plt.figure(figsize = (25,4)) 
ax = fig.add_subplot(111) 
ax.plot(np.arange(0,len(agg_df['Income'])),agg_df['Income']) 
plt.xticks(np.arange(0,len(agg_df['Income'])),np.unique(df['State']),rotation = 90,fontsize = 8) 
ax.set_xlim(0, len(agg_df['Income']))
ax.set_ylabel('Dolars')
ax.set_title('Median household income accumulated by state', fontsize=12)

plt.show()
```
